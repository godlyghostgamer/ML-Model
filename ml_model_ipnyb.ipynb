## Apply `dropna()`

### Subtask:
Show how to remove rows with missing values.


**Reasoning**:
Create a copy of the DataFrame, introduce missing values, and display the DataFrame with missing values.



```python
import numpy as np

# Create a copy of the DataFrame m_df
df_with_na = m_df.copy()

# Introduce missing values (NaN)
# Assign NaN to the cell at row index 1, column 'colA'
df_with_na.loc[1, 'colA'] = np.nan
# Assign NaN to the cell at row index 2, column 'colB'
df_with_na.loc[2, 'colB'] = np.nan

print("DataFrame after introducing missing values:")
display(df_with_na)
```

**Reasoning**:
Remove rows with missing values from the DataFrame and display the resulting DataFrame.



```python
# Apply dropna() to remove rows with missing values
df_dropped_na = df_with_na.dropna()

print("\nDataFrame after removing rows with missing values:")
display(df_dropped_na)
```

## Apply `drop`

### Subtask:
Demonstrate how to drop specific rows or columns.


**Reasoning**:
Create a copy of the m_df DataFrame, drop the row with index 1, drop the column 'colB', and display the result after each drop operation.



```python
# Create a copy of the m_df DataFrame to avoid modifying the original
df_copy = m_df.copy()

# Drop the row with index 1
df_copy_row_dropped = df_copy.drop(1)
print("DataFrame after dropping row with index 1:")
display(df_copy_row_dropped)

# Drop the column named 'colB' from the copied DataFrame
df_copy_col_dropped = df_copy.drop(columns=['colB'])
print("\nDataFrame after dropping column 'colB':")
display(df_copy_col_dropped)
```

## Perform indexing

### Subtask:

**Reasoning**:
Select the row with index 0 from the `m_df` DataFrame using label-based indexing (`.loc`).



```python
# Select the row with index 0
first_row_loc = m_df.loc[0]
print("Row with index 0:")
display(first_row_loc)
```

**Reasoning**:
Select the value in the first row and the column 'colA' from the `m_df` DataFrame using label-based indexing (`.loc`) and select a slice of rows from index 0 to 1 (inclusive) and the column 'colB' from the `m_df` DataFrame using label-based indexing (`.loc`).



```python
# Select the value in the first row and the column 'colA'
value_first_row_colA = m_df.loc[0, 'colA']
print("\nValue in the first row and 'colA' column:")
display(value_first_row_colA)

# Select a slice of rows from index 0 to 1 (inclusive) and the column 'colB'
slice_rows_0_to_1_colB = m_df.loc[0:1, 'colB']
print("\nSlice of rows 0 to 1 (inclusive) and 'colB' column:")
display(slice_rows_0_to_1_colB)
```

## Perform slicing

### Subtask:
Demonstrate how to select a range of data.


**Reasoning**:
Demonstrate how to select a range of data using slicing with both integer-location and label-based methods as requested in the instructions.



```python
# 1. Select a slice of rows from the beginning up to index 2 (exclusive) and all columns using integer-location based slicing (iloc)
slice_iloc_rows = m_df.iloc[:2, :]
print("Slice of rows from beginning up to index 2 (exclusive) and all columns (iloc):")
display(slice_iloc_rows)

# 2. Select a slice of all rows and columns from 'colA' up to 'colB' (inclusive) using label-based slicing (loc)
slice_loc_cols = m_df.loc[:, 'colA':'colB']
print("\nSlice of all rows and columns from 'colA' up to 'colB' (inclusive) (loc):")
display(slice_loc_cols)

# 3. Select a slice of rows from index 1 onwards and the column 'colB' using integer-location based slicing (iloc)
slice_iloc_rows_cols = m_df.iloc[1:, 1]
print("\nSlice of rows from index 1 onwards and column 'colB' (iloc):")
display(slice_iloc_rows_cols)
```
